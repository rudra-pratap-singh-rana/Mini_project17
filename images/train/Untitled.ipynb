```python
import glob
import xml.etree.ElementTree as ET

labels = set()
for xml_file in glob.glob("*.xml"):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    for member in root.findall('object'):
        labels.add(member.find('name').text)

print("Labels found in XML files:", labels)
```

```python
import os
import glob
import xml.etree.ElementTree as ET

# Correction dictionary
label_map = {
    'hello': 'Hello',
    'Hello': 'Hello',
    'yes': 'Yes',
    'Yes': 'Yes',
    'no': 'No',
    'No': 'No',
    'thanks': 'Thank You',
    'Thank You': 'Thank You',
    'I am Utkarsh': 'I am Utkarsh'
}

for xml_file in glob.glob("*.xml"):
    tree = ET.parse(xml_file)
    root = tree.getroot()
    changed = False
    for member in root.findall('object'):
        old_label = member.find('name').text
        if old_label in label_map and member.find('name').text != label_map[old_label]:
            member.find('name').text = label_map[old_label]
            changed = True
    if changed:
        tree.write(xml_file)
        print(f"✅ Fixed labels in {xml_file}")
```
